```python
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = load_iris()
X = df.data #array
y = df.target # array
```

```python
means = np.mean(X, axis=0)
stds = np.std(X, axis=0)
standardized_X = (X - means) / stds
```

```python
perplexity = 30 # Set to 5-50 in general
tsne = TSNE(perplexity=perplexity, n_components=2, init='pca', random_state=919).fit_transform(standardized_X)
df_tsne = pd.DataFrame(tsne, columns=['t1', 't2'])
```

```python
# Plot
plt.scatter(df_tsne.iloc[:, 0], df_tsne.iloc[:, 1], c='black')
plt.xlabel(df_tsne.columns[0])
plt.ylabel(df_tsne.columns[1])
plt.show()
```

```python
# Color Map with target variable
sc = plt.scatter(df_tsne.iloc[:,0],df_tsne.iloc[:,1], vmin=np.min(y), vmax=np.max(y), c=y, cmap=cm.cividis)
plt.colorbar(sc)
plt.title("tSNE with color(target value)")
plt.show()
```
