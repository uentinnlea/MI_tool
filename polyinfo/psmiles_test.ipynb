Draw options

```python
# add draw options
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.drawOptions.addAtomIndices = True
IPythonConsole.drawOptions.addBondIndices= False
IPythonConsole.drawOptions.minFontSize = 16
IPythonConsole.molSize = (600, 200)   # Change image size
IPythonConsole.ipython_useSVG = True  # Change output to SVG

sm = "[*]CC(C)([*])C(=O)OCCOP(=O)([O-])OCC[N+](C)(C)C"
display(PS(sm))
display(PS(sm).canonicalize)
```

Create an alternating copolymer

```python
from psmiles import PolymerSmiles as PS
ps1 = PS('[*]CC[*]')
ps2 = PS('[*]CCO[*]')
ps1.alternating_copolymer(ps2, [0,0])
```

Create block, gradient and random copolymers

```python
display(ps1, ps2)
```

```python
pattern = 'A' * 3 + 'B' * 2 + 'A' * 1
ps1.linear_copolymer(ps2, pattern=pattern)
```

```python
pattern = 'AAABB'
ps1.linear_copolymer(ps2, pattern)
```
