```python
import networkx as nx
import matplotlib.pyplot as plt
```

Creating a graph

```python
g1 = nx.Graph()
g2 = nx.DiGraph()
g3 = g2.to_undirected()
```

```python
type(g1), type(g2), type(g3)
```

Nodes

```python
g = nx.Graph()
g.add_node(2)
nx.draw(g)
```

```python
G = nx.Graph()
G.add_nodes_from([2, 3])
nx.draw(G)
```

```python
G = nx.Graph()
G.add_nodes_from([(4, {"color": "red"}), (5, {"color": "green"})])
nx.draw(G)
```

```python
H = nx.path_graph(10)
G = nx.Graph(H)
display(nx.draw(H))
```

```python
display(nx.draw(G))
```

```python
G = nx.complete_graph(5)
nx.draw(G)
```

Edges

```python
G = nx.Graph()
G.add_edge(1, 2)
nx.draw(G)
```

```python
G = nx.Graph()
G.add_edge(1, 2)
e = (1, 3)
G.add_edge(*e)
nx.draw(G)
```

```python
G.number_of_edges()
```

```python
G.number_of_nodes()
```

Examination

```python
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3)])
G.add_node("hello")
G.add_nodes_from("hello")
G.add_edges_from([(2, 4), (2, 5)])
G.add_edge(2, 'l')
nx.draw(G, with_labels=True)
```

```python
list(G.nodes)
```

```python
list(G.edges)
```

```python
list(G.adj[2])
```

```python
G.degree[2]
```

```python
G.edges([3, 1, 2])
```

```python
G.degree([3, 1, 2])
```

```python
H = G.to_directed()
nx.draw(H, with_labels=True)
plt.show()
```
